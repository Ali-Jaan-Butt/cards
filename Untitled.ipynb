```python
import requests
from scrapy.http import TextResponse
import json
from scrapy import Selector
```

```python
headers = {
'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36'
}
```

```python
res = requests.get('https://www.hareruyamtg.com/en/products/search?product=&category=1&cardset=&colorsType=0&cardtypesType=0&subtype=&format=&illustrator=&foilFlg%5B%5D=0&language%5B%5D=1&stock=1&search=%E6%A4%9C%E7%B4%A2&page=1',headers=headers)
```

```python
response = TextResponse(res.url, body=res.text, encoding='utf-8')
```

```python
ids = json.loads(response.xpath('//ul[@class="itemListLine itemListLine--searched"]/@data-product').extract_first())
```

```python
payload = {}
for i in range(len(ids)):
    payload['product_sub_classes['+str(i)+'][id]'] = ids[i]['id']
    payload['product_sub_classes['+str(i)+'][languageId]'] = ids[i]['languageId']
    payload['product_sub_classes['+str(i)+'][highPriceCode]'] = ids[i]['highPriceCode'] if ids[i]['highPriceCode'] is not None else ''
payload['css'] = 'itemList'
payload['query[category]'] = 1
payload['query[foilFlg][]'] = 0
payload['query[language][]'] = 1
payload['query[stock]'] = 1
payload['query[search]'] = '検索'
payload['query[page]'] = 1
```

```python
response = requests.post('https://www.hareruyamtg.com/en/products/search/lazy',headers=headers,data=payload)
```

```python
sel = Selector(text=response.text)
```

```python
sel.xpath('//ul[@class="itemListLine itemListLine--searched"]/li')
```
